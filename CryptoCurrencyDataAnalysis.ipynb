## Data Preparation and Cleaning

**TODO** - I changed the data format into date_time format for manipulating with dates.I also changed the dat types of some of the columns into float for performing arthimetic.I also added some usefull colums such as open-close-variation , high-low variation and Average coin price per day.



```python
crypto_df
```

```python
crypto_df['Date'] = pd.to_datetime(crypto_df.Date)
```

```python
crypto_df
```

```python
crypto_df.describe()
```

```python
crypto_df.Currency.unique()
```

```python
crypto_df.info()
```

```python
crypto_df['Open'] = pd.to_numeric(crypto_df['Open'],errors = 'coerce')
crypto_df['Close'] = pd.to_numeric(crypto_df['Close'],errors = 'coerce')
crypto_df['High'] = pd.to_numeric(crypto_df['High'],errors = 'coerce')
crypto_df['Low'] = pd.to_numeric(crypto_df['Low'],errors = 'coerce')
#crypto_df['Volume'] = pd.to_numeric(crypto_df['Volume'],errors = 'coerce')
#crypto_df['Market Cap'] = pd.to_numeric(crypto_df['Market Cap'],errors = 'coerce')
```

```python
crypto_df['Open_close_variation'] = crypto_df['Open'] - crypto_df['Close']
```

```python
crypto_df
```

```python
crypto_df['High_Low_variation'] = crypto_df['High'] - crypto_df['Low']
```

```python
crypto_df
```

```python
crypto_df.loc[1480]
```

```python
crypto_df['Average_Coin_Price'] = (crypto_df['High'] + crypto_df['Low'] + crypto_df['Open'] + crypto_df['Close'])/4
```

```python
crypto_df
```

```python
crypto_df['year'] = pd.DatetimeIndex(crypto_df.Date).year
crypto_df['month'] = pd.DatetimeIndex(crypto_df.Date).month
crypto_df['day'] = pd.DatetimeIndex(crypto_df.Date).day
crypto_df['weekday'] = pd.DatetimeIndex(crypto_df.Date).weekday
```

## Asking and Answering Questions

TODO - I asked the question which is mostly asked and researched while purchasing or selling crypto currency.



#### Q1: TODO - What is th maximum average coin prices of the currenceies?

```python
crypto_currency_max_df = crypto_df.groupby(['Currency'])[['Average_Coin_Price']].max()
crypto_currency_max_df
```

#### Q2: TODO -What is th minimum average coin prices of the currenceies?

```python
crypto_currency_min_df = crypto_df.groupby(['Currency'])[['Average_Coin_Price']].min()
crypto_currency_min_df
```

#### Q3: TODO - Which coin could  be able to give maximum profit? 

```python
crypto_currency_profit_df=crypto_currency_max_df-crypto_currency_min_df
crypto_currency_profit_df.max()
```

#### Q4: TODO -  Which coin could  be able to give minimum profit? 

```python
crypto_currency_profit_df=crypto_currency_max_df-crypto_currency_min_df
crypto_currency_profit_df.min()
```

#### Q5: TODO - What are the maximum Market Cap?

```python
crypto_currency_market_df = crypto_df.groupby(['Currency'])[['Market Cap']].max()
crypto_currency_market_df
```
